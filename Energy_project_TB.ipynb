```python
# This workbook analyses world wide Energy consumtion data
import pandas as pd
import matplotlib.pyplot as plt
#import panda and matplotlib
```

```python
raw = pd.read_csv("https://raw.githubusercontent.com/owid/energy-data/master/owid-energy-data.csv")
# import WW energy consumption data file from source
```

```python
pd.set_option('display.max_columns', None) 
raw.head(17240)
#display all col headers so I can select the ones I need
# see top and bottom of the dataset so I can see the years covered.
```

```python
raw.shape
# see the fata set size
```

```python
ire = raw[raw['country'] == 'Ireland']
ire.head(130)
# define a dataframe for Ireland only
```

```python
ren_energy_consumption_ire = ire[["year", "other_renewable_consumption", "hydro_consumption", "solar_consumption", "wind_consumption"]]
year_range = ren_energy_consumption_ire["year"].isin(range(1990,2021))
ren_energy_consumption_ire = ren_energy_consumption_ire[year_range]
ren_energy_consumption_ire.head(130)
# Define a dataframe for Irelands Renewable energy consumption from 1990 to 2020
```

```python
ren_energy_consumption_ire.plot(x='year', ylabel='Terrawatt Hour', kind='bar', stacked=True,figsize=(20,6),
        title='Irish Renewable Consumption 1990 - 2020')
plt.show()
# show Irelands growth in renewable Energy consumption from 1990 to 2020
# Findings: Irelands steady growth in renewable energy has primarily been driven by the gorwth in Wind Energy Consumption since 2005
# Hydro has remained consistent, solar is very small and other sources has begun to grow since 2007
```

```python
energy_consumption_ire = ire[["year", "other_renewable_consumption", "hydro_consumption", "solar_consumption", "wind_consumption", "primary_energy_consumption"]]
year_range2 = energy_consumption_ire["year"].isin(range(1990,2021))
energy_consumption_ire = energy_consumption_ire[year_range2]
column_names = ["other_renewable_consumption", "hydro_consumption", "solar_consumption", "wind_consumption"]
energy_consumption_ire['renew_total']= energy_consumption_ire[column_names].sum(axis=1)
energy_consumption_ire.head(130)
# df of total Energy consumption v total reneable consumption in irelnd
# totalise Renewable consumption
```

```python
fig, ax = plt.subplots()
x = energy_consumption_ire['year']
y1 = energy_consumption_ire['primary_energy_consumption']
y2 = energy_consumption_ire['renew_total']
ax.plot(x,y1)
ax.plot(x,y2)
ax.set_xlabel('Year')
ax.set_ylabel('Energy Consumption (twh)')
ax.set_title('Ireland Totoal Primary energy Consumption v Renewable Energy Consumption')
plt.show()
# Line plot Irelands primary Energy Consumtion and renewable consumption from 1990 to 2020.
# Findings: 1) Irelands Enegy Consuption grew strongly between 1990 and 2000 but with some flutctuations has levelled since then
# 2) Irelands renewable energy consumption has grown consistently since 2005 and has therefore displaced some fossil fuel consumption
```

```python
raw = pd.read_csv("https://raw.githubusercontent.com/owid/energy-data/master/owid-energy-data.csv")
ren_ww = raw[['year', "country", "other_renewable_consumption", "hydro_consumption", "solar_consumption", "wind_consumption", "population", "gdp", "energy_per_capita", "primary_energy_consumption"]]
year_range_ww = ren_ww["year"].isin(range(2019,2020))
ren_ww = ren_ww[year_range_ww]
ren_ww = ren_ww.fillna(0)
ren_ww.head(1000)
# dataframe with the renewable data for year 2019 for all countries & population, gdp and energy per capita, primary Energy consumption
# choose 2019 as recent & less missing data than 2020
# all NaN values set to zero
```

```python
indexNames = ren_ww[(ren_ww['country'] == 'World') ].index
ren_ww.drop(indexNames , inplace=True)
column_names = ["other_renewable_consumption", "hydro_consumption", "solar_consumption", "wind_consumption"]
ren_ww['renew_total']= ren_ww[column_names].sum(axis=1)
ren_ww.head()
# exclude world as it is an aggregate of all countries
# sum renewable energy columns to new col renew_total
```

```python
import seaborn as sns
sns.histplot(ren_ww['energy_per_capita']).set(title='2019 - per capita Energy Consumption - 2019')
# undestand the distribution of Energy consumption an a per capita basis
# Histogram of the Energy per capita, using Seaborn
# majority of countries use low levels of Energy, therefore if they develop towards the energy consumption of higher Nations there will be further pressure on energy resources.
```

```python
high_energy_users = ren_ww[['country', 'energy_per_capita']]
pd.set_option('display.max_rows', None)
high_energy_users.sort_values(by=['energy_per_capita'], ascending=False)

# identify the Highest countries per captia usage by listing in order of usage per capita (starting with the highest)
# Findings: 1) Oil rich and very hot middle eastern states amongst the highest per capita users
#           2) Colder developed countries such as canada, scandinavian countries, Northern Europe are also high.
# Note the list of "Countires" includes some blocks for comparison e.g Eu 27, OECD 38 etc..
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.scatterplot(data=ren_ww, x='renew_total', y='primary_energy_consumption', size='energy_per_capita', palette='deep', hue='gdp', markers="country", sizes=(20, 600), hue_norm=(0, 10), alpha=.5)
#Use a scatter plot to understand the relationships of enrgy consumption (renewable & overall), GDP, per Capita consumption.
#Use Seaborn to code, using maker size and hue to give 4 variables
#findings 1) No Clear pattern emerges as too many other factors such as climate, energy resources etc... are signifcant factors
```

```python
high_energy_users = ren_ww[['country', 'primary_energy_consumption']]
pd.set_option('display.max_rows', None)
high_energy_users.sort_values(by=['primary_energy_consumption'], ascending=False)
# list the countires & Blocks in orcer of Primary energy consumption. Primary Energy consumption is a term that look for as example at the source of eletric generation
# as primary energy.
# Finding: The largest industrialised "Developed" countires are the largest consumers of Energy.
# China is the largest single contry user despite moderate usage per capita.
```

```python
high_energy_users = ren_ww[['country', 'renew_total']]
pd.set_option('display.max_rows', None)
high_energy_users.sort_values(by=['renew_total'], ascending=False)
# list the countires & Blocks in orcer of renewable energy consumption. Note: Renewable energy consumption is not the same as Renewable Energy generation
# Some Contries may import or export Renewable Energy e.g on corss border Electical grids for example.
# Finding: The largest industrialised "Developed" countires are the largest consumers of Renewable Energy.
# China is the largest single contry user.
```

```python
raw = pd.read_csv("https://raw.githubusercontent.com/owid/energy-data/master/owid-energy-data.csv")
world_view = raw[raw['country'] == 'World']
world_view_capita = world_view[["year", "fossil_fuel_consumption", "fossil_energy_per_capita", "low_carbon_consumption", "low_carbon_energy_per_capita", "primary_energy_consumption"]]
year_range4 = world_view_capita["year"].isin(range(1990,2021))
world_view_capita = world_view_capita[year_range4]
world_view_capita.head(40)
# select Worldwode data from 1990 to 2020 for the specified columns
# show the dataframe
```

```python
fig, ax = plt.subplots(2, 1)
x = world_view_capita['year']
y1 = world_view_capita['fossil_fuel_consumption']
y2 = world_view_capita['low_carbon_consumption']
y3 = world_view_capita['fossil_energy_per_capita']
y4 = world_view_capita['low_carbon_energy_per_capita']
ax[0].plot(x,y1)
ax[0].plot(x,y2)
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Energy Consumption (twh)')
ax[0].set_title('WW Fossil energy Consumption & Renewable Energy Consumption')
ax[1].plot(x,y3)
ax[1].plot(x,y4)
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Energy Consumption per capita (kwh)')
ax[1].set_title('WW Fossil energy Cons per cap & Renewable Energy Cons per cap')
plt.show()
# define 2 subplots, one on top of the other to compare absolute & per capita Energy consumption for the WW data 
# using matplotlib
# findings: Consumption of Low carbon and Fossil fuel Energy continues to grow onboths an absolute and per capita basis. 
# The increase on a per capita basis is more modest than absolute in fossilfuel
```
